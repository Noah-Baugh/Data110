```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#setting the array of data the chart will use
data = {
    'Offense': ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
    'Offenses per 100,000 (2023)': [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5],
    'Change in Offenses per 100,000 (2022-2023)': [-3.3, -8.1, -4.9, -12, -12, -9.8, -8]
}

df = pd.DataFrame(data)

#make bars and adjust width
x = np.arange(len(df['Offense']))
width = 0.35

#labelling and adjusting positions
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, df['Offenses per 100,000 (2023)'], width, label='Offenses per 100,000 (2023)')
rects2 = ax.bar(x + width/2, df['Change in Offenses per 100,000 (2022-2023)'], width, label='Change in Offenses per 100,000 (2022-2023)')
ax.set_ylabel('Number of Offenses')
ax.set_title('Offenses per 100,000 People and Change (2022-2023)')
ax.set_xticks(x)
ax.set_xticklabels(df['Offense'])
ax.legend()

# Add gridlines
ax.grid(True)

fig.tight_layout()
plt.show()
```

```python

#Create the colors
colors = plt.cm.get_cmap('tab20').colors[:len(df['Offense'])]

# Create the horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(df['Offense'], df['Change in Offenses per 100,000 (2022-2023)'], color=colors)

# Set x and y
ax.set_xlabel('Percentage Change in Offenses')
ax.set_ylabel('Offense Type')
ax.set_title('Percentage Change in Offense Rates (2022-2023)')

# Add gridlines
ax.grid(True)
plt.show()
```

# why seeing the change over time is more insightful

Seeing the change over time in a graph shows the variance of data over a set interval. It often gives more information than just how much happened. You get to see the change instead of the misleading count. A snapshot doesn't give you the full story, a video will. It's more important to see the variance.
